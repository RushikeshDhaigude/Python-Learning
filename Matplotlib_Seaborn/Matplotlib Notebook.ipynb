# Matplotlib Tutorial

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

## Line Plot

```python
x = np.random.randint(10,50,10)
y = np.random.randint(20,70,10)
```

```python
plt.plot(x,y, color = 'red', marker = '*')
plt.xlabel('Scores')
plt.ylabel('Age')

plt.title('Score Vs Age')
```

```python
day=[1,2,3,4,5,6,7]

vivo_sales=[56000,34000,45000,30000,70000,60000,89000]

oppo_sales=[46000,74000,65000,90000,20000,34000,49000]

plt.figure(figsize=(15,6))

plt.plot(day,vivo_sales,color="red",marker="*",label="vivo")

for i,j in zip(day,vivo_sales):
    plt.text(i,j+1,"({},{})".format(i,j))

plt.plot(day,oppo_sales,color="blue",marker="*",label="oppo")

for i,j in zip(day,oppo_sales):
    plt.text(i,j+1,"({},{})".format(i,j))

plt.xlabel("day")
plt.ylabel("sales")
plt.title("day vs sales")
plt.legend()
plt.show()
```

## Grouped Bar Chart

```python
Python_marks= [70,86,76,78]
R_marks= [67,88,65,78]

index = np.arange(4)

plt.bar(x = index, height = Python_marks, width=0.4, label = 'Python')
plt.bar(x= index+0.4, height = R_marks, width = 0.4, label = 'R')

plt.xticks(ticks = index+ (0.4/2), labels = ('A','B','C','D'))  

plt.legend()
```

## Stacked Chart

```python
Python_marks= [70,86,76,78]
R_marks= [67,88,65,78]

index = np.arange(4)
```

```python
plt.bar(x = index, height = Python_marks, label = 'Python')
plt.bar(x = index, height = R_marks, bottom = Python_marks, label = 'R')

plt.legend()

plt.xticks(ticks = index, labels = ('A','B','C','D')) 
```

```python
python_marks=[56,67,89,48,60]
R_marks=[45,80,40,87,59]

index=np.arange(5)

plt.bar(x=index,height=python_marks,width=0.5,label="python")

plt.bar(x=index,height=R_marks,bottom=python_marks,width=0.5,label="R")

plt.legend()

plt.xticks(ticks=index,labels=("A","B","C","D","E"))
```
